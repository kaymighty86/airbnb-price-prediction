```python
#Step 14: Treating Outliers - Only for the continuous numerical data
#We intend to deal with our outliers by using our interquartile range
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['cleanliness_rating'])
```

```python
# Step 14: Treating Outliers
for i in ['cleanliness_rating', 'guest_satisfaction_overall', 'bedrooms', 'dist', 'metro_dist', 'attr_index', 'attr_index_norm', 'rest_index','rest_index_norm', 'lng', 'lat']:
    lw, uw=wisker(df[i])
    df[i]=np.where(df[i]<lw, lw, df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
# Step 14: Treating Outliers
# We intend to plot a boxplot to confirm if we no longer have outliers
for i in ['cleanliness_rating', 'guest_satisfaction_overall', 'bedrooms', 'dist', 'metro_dist', 'attr_index', 'attr_index_norm', 'rest_index','rest_index_norm', 'lng', 'lat']:
    sns.boxplot(df[i])
    plt.show()
```

#### Data Encoding and Scaling

**Encoding**

```python
df_copy = df.reset_index(drop=True) #we're copying the dataframe before begining transforming
```

```python
#seperate target from features
x = df_copy.drop("realSum", axis=1)
y = df_copy["realSum"]
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

#OneHot encoding
# room_type_encoded = pd.get_dummies(x["room_type"], prefix = "room_type").astype(int)
# room_shared_encoded = pd.get_dummies(x["room_shared"], prefix = "room_shared").astype(int)
# room_private_encoded = pd.get_dummies(x["room_private"], prefix = "room_private").astype(int)
# superhost_encoded = pd.get_dummies(x["host_is_superhost"], prefix = "host_is_superhost").astype(int)

label_encoder = LabelEncoder()
room_type_encoded = pd.DataFrame(label_encoder.fit_transform(x["room_type"]), columns=["room_type"])
room_shared_encoded = pd.DataFrame(label_encoder.fit_transform(x["room_shared"]), columns=["room_shared"])
room_private_encoded = pd.DataFrame(label_encoder.fit_transform(x["room_private"]), columns=["room_private"])
superhost_encoded = pd.DataFrame(label_encoder.fit_transform(x["host_is_superhost"]), columns=["host_is_superhost"])
```

```python
room_type_encoded
```

Merging the encoded columns back

```python
labels = ["room_type", "room_shared", "room_private", "host_is_superhost"]

x_encoded = x.drop(labels, axis=1)#drop the existing columns first
x_encoded = pd.concat([x_encoded, room_type_encoded, room_shared_encoded, room_private_encoded, superhost_encoded], axis=1) #join the new columns
```

### Working on the Baseline Model

**Split Train and Test Datasets**

```python
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(25)

x_train, x_test, y_train, y_test = train_test_split(x_normalised, y, test_size=0.2)

x_train.head()
```

```python
y_train.tail()
```

**Model Fitting**

Establishing our evaluation metrics

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

#A reusable evaluation function
def evaluate(y_test, y_pred):
    return pd.Series({
        "r2": r2_score(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "MSE": mean_squared_error(y_test, y_pred),
        "RMSE": root_mean_squared_error(y_test, y_pred),
    })
```

Fitting the multiple baseline models and evaluating each

```python
from sklearn.ensemble import RandomForestRegressor

rgr_baseline = RandomForestRegressor(n_estimators=100)
rgr_baseline.fit(x_train, y_train)
y_pred = rgr_baseline.predict(x_test)
evaluate(y_test, y_pred)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)
evaluate(y_test, y_pred_ridge)
```

```python
from sklearn.linear_model import LinearRegression

linearReg = LinearRegression()
linearReg.fit(x_train, y_train)
y_pred_linReg = linearReg.predict(x_test)
evaluate(y_test, y_pred_linReg)
```

### Feature Selection

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import Lasso

best_model = RFECV(linearReg, cv = 5, scoring="r2", n_jobs=-1, verbose = 2)
best_model.fit(x_train, y_train)
y_pred_fs = best_model.predict(x_test)
evaluate(y_test, y_pred_fs)
```

### Hyper-Parameter Tunning

Using Random Search CV

```python
linearReg.get_params()
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    'copy_X': [True, False],
    'fit_intercept': [True, False],
    'positive': [True, False],
    'tol': [1e-06]
 }
```

```python
tuned_model = RandomizedSearchCV(linearReg, param_distributions = params, cv = 5, n_iter=15, n_jobs= -1, verbose = 2)
tuned_model.fit(x_train, y_train)
y_pred_tuned = tuned_model.predict(x_test)
evaluate(y_test, y_pred_tuned)
```
